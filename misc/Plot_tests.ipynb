```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cuda_obj = np.array([438, 457, 545, 549, 602, 1177, 2997])
cpp_obj  = np.array([51173, 55496, 83236, 114733, 168045, 339067, 629421])
x_obj = np.array([1,2,5,10,20,50,100])

cuda_res = np.array([71, 139, 260, 466, 1042, 2373, 4232, 9555])
cpp_res  = np.array([3617, 9781, 22082, 38761, 87417, 194921, 341254, 765919])
x_res = np.array([144, 240, 360, 480, 720, 1080, 1440, 2160])

cuda_spp = np.array([19, 20, 40, 69, 139, 256, 475, 788, 1291, 1911, 2268])
cpp_spp = np.array([943, 1032, 2331, 4647, 9325, 19227, 37348, 57676, 92821, 139513 ,186563])
x_spp = np.array([1, 10, 25, 50, 100, 200, 400, 600, 1000, 1500, 2000])
```

```python
y1, y2 = cuda_spp, cpp_spp
x = x_spp

fig, ax1 = plt.subplots(dpi = 200)
ax2 = ax1.twinx()

a1 = ax1.plot(x, y1/1000, label='CUDA')
a2 = ax2.plot(x, y2/1000, 'r', label='C++')

ax1.set_ylabel('Tempo de execução (s)')
ax2.set_ylabel('Tempo de execução (s)', rotation=-90, labelpad=10)

ax1.set_xlabel('Amostras por pixel')

lns = a1+a2
labs = [l.get_label() for l in lns]

ax1.legend(lns, labs)

ax1.set_title('Tempo de execução em função do número de amostras por pixel')
```

```python
y1, y2 = cuda_res, cpp_res
x = x_res

fig, ax1 = plt.subplots(dpi = 200)
ax2 = ax1.twinx()

a1 = ax1.plot(x, y1/1000, label='CUDA')
a2 = ax2.plot(x, y2/1000, 'r', label='C++')

ax1.set_ylabel('Tempo de execução (s)')
ax2.set_ylabel('Tempo de execução (s)', rotation=-90, labelpad=10)

ax1.set_xlabel('Resolução (p)')

lns = a1+a2
labs = [l.get_label() for l in lns]

ax1.legend(lns, labs)

ax1.set_title('Tempo de execução em função da resolução')
```
